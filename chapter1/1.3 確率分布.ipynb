# 離散の場合
## ポアソン分布
$$ P(Z=k)= \frac{\lambda^k e^{-\lambda}}{k!} $$

特徴
- 期待値が$\lambda$と等しい
- $\lambda$ を大きくすれば大きな値が出る確率が高くなる
- 確率変数は非負

```python
# poisson distribution
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
%matplotlib inline
figsize(12.5, 4)

import scipy.stats as stats
import numpy as np
poi = stats.poisson
lambda_ = [4.25, 18]
colors = ["#348ABD", "#A60628"]

a = np.arange(16)
plt.bar(a, poi.pmf(a, lambda_[0]), color=colors[0],
        label="$\lambda = %.1f$" % lambda_[0], alpha=0.60,
        edgecolor=colors[0], lw="3")

plt.bar(a, poi.pmf(a, lambda_[1]), color=colors[1],
        label="$\lambda = %.1f$" % lambda_[1], alpha=0.60,
        edgecolor=colors[1], lw="3")

plt.xticks(a, a)
plt.legend()
plt.ylabel("probability of $k$")
plt.xlabel("$k$")
plt.title("Probability mass function of a Poisson random variable; differing \
$\lambda$ values")
```

# 連続の場合\
## 指数分布
$$ f_z(z|\lambda)=\lambda e^{-\lambda z}, z\geq0 $$

特徴
- 期待値が$\frac{1}{\lambda}$
- 確率変数が連続値に対応

```python
# exponential distribution
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5, 1]

for l, c in zip(lambda_, colors):
    plt.plot(a, expo.pdf(a, scale=1. / l), lw=3,
             color=c, label="$\lambda = %.1f$" % l)
    plt.fill_between(a, expo.pdf(a, scale=1. / l), color=c, alpha=.33)

plt.legend()
plt.ylabel("PDF at $z$")
plt.xlabel("$z$")
plt.ylim(0, 1.2)
plt.title("Probability density function of an Exponential random variable;\
 differing $\lambda$")
```
